```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
airline = pd.read_csv("EastWestAirlines.csv")
airline
```

```python
airline.info()
```

```python
airline2 = airline.drop(["ID#"], axis = 1)
airline2
```

```python
airline2_norm = pd.DataFrame(normalize(airline2), columns = airline2.columns)
airline2_norm
```

```python
plt.figure(figsize = (10, 7))
dendograms = sch.dendrogram(sch.linkage(airline2_norm, "complete"))
```

```python
plt.figure(figsize = (10, 7))
dendograms = sch.dendrogram(sch.linkage(airline2_norm, "average"))
```

```python
clusters = AgglomerativeClustering(n_clusters = 5, affinity = "euclidean", linkage = "ward")
clusters
```

```python
y = pd.DataFrame(clusters.fit_predict(airline2_norm), columns = ["clustersID"])
y["clustersID"].value_counts()
```

```python
airline2["clustersID"] = clusters.labels_
airline2
```

```python
airline2.groupby("clustersID").agg(["mean"]).reset_index()
```

```python
plt.figure(figsize = (10, 7))
plt.scatter(airline2["clustersID"],airline2["Balance"], c = clusters.labels_)
```
